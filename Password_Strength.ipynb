```python
password_tuple=np.array(data)
```

```python
password_tuple
```

```python
#Shuffling Data to provide robustness
```

```python
import random
random.shuffle(password_tuple)
```

```python
x=[lables[0] for lables in password_tuple]  #independent data
y=[lables[1] for lables in password_tuple]  #dependent data
```

```python
x
```

```python
#spliiting Into Chraacters

def word_divide_char(inputs):
    character=[]
    for i in inputs:
        character.append(i)
    return character
    
```

```python
word_divide_char('kzde5577')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer(tokenizer=word_divide_char)
```

```python
X=vectorizer.fit_transform(x)
```

```python
X.shape
```

```python
vectorizer.get_feature_names()
```

```python
first_document_vector=X[0]
first_document_vector        #Importance of Each X
```

```python
first_document_vector.T.todense()
```

```python
#Matrix with first column as feature and second column as importance
```

```python
df=pd.DataFrame(first_document_vector.T.todense(),index= vectorizer.get_feature_names(),columns={'TF-IDF'})
df.sort_values(by=['TF-IDF'],ascending=False)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression(random_state=0,multi_class='multinomial')
```

```python
clf.fit(X_train,y_train)
```

```python
dt=np.array(['%@123abcd'])
pred=vectorizer.transform(dt)
clf.predict(pred)
```

```python
y_pred=clf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
